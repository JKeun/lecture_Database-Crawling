# SQL(WHERE, ORDER BY 2개) & pandas(행렬곱) - sakila(db)
* 렌탈가격(rental_rate)이 1달러 이상인 영화제목(title)/렌탈가격을 내림차순으로 뽑아라(film table)

```python
import MySQLdb
import pandas as pd

db = MySQLdb.connect(
    'db.fastcamp.us',
    'root',
    'dkstncks',
    'sakila',
    charset='utf8',
)

cursor = db.cursor()
```

---
### SQL

```python
SQL_QUERY = """
    SELECT 
        film.title,
        film.rental_rate
    FROM 
        film
    WHERE
        film.rental_rate >= 1
    ORDER BY
        2 DESC
    ;
"""

pd.read_sql(SQL_QUERY, db)
```

```python
SQL_QUERY = """
    SELECT 
        film.title,
        film.rental_rate
    FROM 
        film
    WHERE
        film.rental_rate >= 1
    ORDER BY
        2 DESC,
        1 ASC       # default : ASC
    ;
"""

pd.read_sql(SQL_QUERY, db)
```

### pandas

```python
df = pd.read_sql("SELECT * FROM film;", db)

df[df['rental_rate'] >= 1][['title', 'rental_rate']].sort_values('rental_rate', ascending = False)
# df[['title', 'rental_rate']]는 데이터를 받아오는 명령이기때문에 리스트로 묶고
# df['rental_rate']는 boolean Series를 받기위한것이기 때문에 리스트로 묶지않는다
```

```python
df = pd.read_sql("SELECT * FROM film;", db)

df[df['rental_rate'] >= 1][['title', 'rental_rate']].sort_values(['rental_rate', 'title'],
                                                                 ascending = [False, True])
```
