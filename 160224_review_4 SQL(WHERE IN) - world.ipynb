# SQL(WHERE IN) - world
* 대륙이 South America 이거나 North America인 국가, 대륙, 인구를 뽑아라

```python
import MySQLdb
import pandas as pd

db = MySQLdb.connect(
    'db.fastcamp.us',
    'root',
    'dkstncks',
    'world',
    charset='utf8',
)

curosr = db.cursor()
```

---
### SQL

```python
SQL_QUERY = """
    SELECT
        Name,
        Continent,
        Population,
        SurfaceArea
    FROM
        Country
    WHERE
        Continent = 'South America'
    ;        
"""

pd.read_sql(SQL_QUERY, db)
```

```python
SQL_QUERY = """
    SELECT
        Name,
        Continent,
        Population,
        ROUND(SurfaceArea/9363520 * 1000, 0) '미국의 면적이 1000일때 면적',
        ROUND(Population/278357000 * 1000, 0) '미국 인구가 1000명일 때의 인구수',
        ROUND(ROUND(SurfaceArea/9363520 * 1000, 0)/ROUND(Population/278357000 * 1000, 0),2) '면적당 인구수'
    FROM
        Country
    WHERE
        Continent IN ('South America', 'North America')
    ORDER BY
        2,
        3 DESC
    ;
"""

pd.read_sql(SQL_QUERY, db)
```
