# WHERE LIKE
* Country 테이블에서 GovernmentForm에 "Republic"이란 텍스트가 포함된 모든 row를 뽑아라

```python
import MySQLdb
import pandas as pd

db = MySQLdb.connect(
    'db.fastcamp.us',
    'root',
    'dkstncks',
    'world',
    charset='utf8',
)

cursor = db.cursor()
```

---
### SQL

```python
SQL_QUERY = """
    SELECT
        Name, GovernmentForm
    FROM
        Country
    WHERE GovernmentForm LIKE "%Republic%"
    ;
"""
# % => 모든 텍스트 ( 없어도 되고 있어도 된다 )
# 앞에만 있으면 Republic으로 시작되는 텍스트를 뽑음

pd.read_sql(SQL_QUERY, db)
```

### pandas

```python
df = pd.read_sql("SELECT * FROM Country;", db)

df = df[['Name', 'GovernmentForm']]

df[df['GovernmentForm'].str.contains('Republic')]
```

# SQL(DISTINCT)
* unique한 값들만 뽑기

---
### SQL

```python
SQL_QUERY = """
    SELECT
        DISTINCT(GovernmentForm)
    FROM
        Country
    ;
"""

pd.read_sql(SQL_QUERY, db)
```

### pandas

```python
df = pd.read_sql("SELECT * FROM Country;", db)

df['GovernmentForm'].unique()
```

### Python

```python
# set() => 집합 : unique값 find해줌
# list() => 집합(set)

print(type(set(df['GovernmentForm'])), type(list(set(df['GovernmentForm']))))

list(set(df['GovernmentForm']))
```
