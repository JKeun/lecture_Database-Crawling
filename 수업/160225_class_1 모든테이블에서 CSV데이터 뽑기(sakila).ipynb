```python
import pandas as pd
import MySQLdb

db = MySQLdb.connect(
    "db.fastcamp.us",
    "root",
    "dkstncks",
    "sakila",
    charset='utf8',
)

cursor = db.cursor()
```

```python
SQL_QUERY = """
    SHOW TABLES;
"""

df = pd.read_sql(SQL_QUERY, db)
df
```

```python
df.head(3)
```

```python
df["SQL_QUERY"] = df["Tables_in_sakila"].apply(
lambda x: "SELECT * FROM {table_name}".format(table_name=x)
)
# lambda는 변수를 받지않는 함수
# apply라는 함수의 변수역할을 하는 함수
```

```python
df
```

```python
for index, row in df.iterrows():
    SQL_QUERY = row["SQL_QUERY"]
    print(SQL_QUERY)
```

```python
for index, row in df.iterrows(): # 데이터프레임을 for문돌릴때 기본 # iterrows는 인덱스, 로우 튜플형태로 나오기때문에 변수2개넣어준것
    TABLE_NAME = row["Tables_in_sakila"]
    SQL_QUERY = row["SQL_QUERY"]
    print(SQL_QUERY)
    
    # DataFrame을 새롭게 만든다
    table_df = pd.read_sql(SQL_QUERY, db)
    table_df.to_csv("{table_name}.csv".format(table_name=TABLE_NAME))
```
