```python
response = requests.get("http://www.naver.com/")
```

```python
response.content
```

```python
# DOM => Document Object Model (html을 구조화시키는 역할)
dom = BeautifulSoup(response.content, "html.parser")
dom
```

```python
#realrank라는 id를 가진 html element를 찾아서, 그 안에 list를 뽑아옴
# id는 유니크함(1개뿐), 'up'이라는 class를 가진 것을 뽑아와라 

# html element => 그냥 적는다(ol)
# id => '#'(realrank)
#  class => "."(up)


dom.select('ol#realrank li.up')
```

```python
real_rank_elements = dom.select("ol#realrank li.up") # li ~ 다음 li전까지임

for real_rank_element in real_rank_elements:
    attrs = real_rank_element.select_one("a").attrs # a href~에 들어있는 모든 데이터 다가져옴
#    print(attrs)
    keyword = attrs.get("title")
    print(keyword)
# response할떄의 시점의 데이터들임
```

```python
real_rank_elements = dom.select("ol#realrank li.up") # li ~ 다음 li전까지임

for real_rank_element in real_rank_elements[:-1]: #네이버의 1위가 두번나오는 오류 제거해주기위해
    attrs = real_rank_element.select_one("a").attrs # a href~에 들어있는 모든 데이터 다가져옴
#    print(attrs)
    keyword = attrs.get("title")
    print(keyword)
```

```python
real_rank_elements = dom.select("ol#realrank li.up") # li ~ 다음 li전까지임

for real_rank_element in real_rank_elements[:-1]:
    attrs = real_rank_element.select_one("a").attrs # a href~에 들어있는 모든 데이터 다가져옴
#    print(attrs)
    keyword = attrs.get("title")
    link = attrs.get('href')
    print(keyword, link)
```

## 정적인 웹페이지 크롤링수업 끝
* 블로그 제목 크롤링하기
    * 검색어 : 신사동 오피스텔

```python
response = requests.get("https://search.naver.com/search.naver?where=nexearch&query=%EC%8B%A0%EC%82%AC%EB%8F%99+%EC%98%A4%ED%94%BC%EC%8A%A4%ED%85%94&sm=top_hty&fbm=1&ie=utf8")
```

```python
dom = BeautifulSoup(response.content, "html.parser")
dom
```

```python
# type01안에 리스트들이 어떤 공통적인 특징을 갖고있는가?
# 바로 type01안에 있는 sh_blog_top이라는 특징
# => ul.type01 li.sh_blog_top
# (ul.class1.class2 클래스가 여러개일경우는)

post_elements = dom.select("ul.type01 li.sh_blog_top")
post_elements
```

```python
for post_element in post_elements:
    title_element = post_element.select_one("a.sh_blog_title") #a다음class sh_blog~임
    title = title_element.attrs.get('title')
    print(title)
```
