### 이게 어제 과제1의 답안임
* pandas로 db에 저장시 덮어쓰기 때문에 위험
* 직접 db에서 테이블을 만들고 그 형식에 맞춰 집어넣어야함

```python
for index, row in df.iterrows():
    address = row.get("address")
    # address = row("address")
    phonenumber = row.get("phonenumber")
    deposit = row.get("deposit")
    rent = row.get("rent")
    
    SQL_QUERY = """
        INSERT INTO zigbang (address, phonenumber, deposit, rent)
        VALUES (
            '{address}',
            '{phonenumber}',
            '{deposit}',
            '{rent}'
            )
            ;
        """.format(
                address=address,
                phonenumber=phonenumber,
                deposit=deposit,
                rent=rent,
        )
    
    print(SQL_QUERY)
```

### 시퀄프로 들어가서 db 생성

```python
import MySQLdb

db = MySQLdb.connect(
    "db.fastcamp.us",
    "root",
    "dkstncks",
    "jkeun_sample",
    charset='utf8',
)

cursor = db.cursor()
```

```python
for index, row in df.iterrows():
    address = row.get("address")
    # address = row("address")
    phonenumber = row.get("phonenumber")
    deposit = row.get("deposit")
    rent = row.get("rent")
    
    SQL_QUERY = """
        INSERT INTO zigbang (address, phonenumber, deposit, rent)
        VALUES (
            '{address}',
            '{phonenumber}',
            '{deposit}',
            '{rent}'
            )
            ;
        """.format(
                address=address,
                phonenumber=phonenumber,
                deposit=deposit,
                rent=rent,
        )
    
    
    cursor.execute(SQL_QUERY)
    db.commit()
```
