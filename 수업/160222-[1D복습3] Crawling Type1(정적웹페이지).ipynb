```python
real_rank_elements = dom.select('ol#realrank li.up')

for real_rank_element in real_rank_elements:
    attrs = real_rank_element.select_one('a').attrs
    # a로 시작하는 list안의 모든 데이터를 가져옴
    print(attrs)
```

* **title만 뽑아오기**

```python
real_rank_elements = dom.select('ol#realrank li.up')

for real_rank_element in real_rank_elements:
    attrs = real_rank_element.select_one('a').attrs
    # a herf...에 들어있는(<a list>안의) 모든 데이터를 가져옴
    # print(attrs)
    keyword = attrs.get("title")
    # attrs에서 "title"다음의 데이터를 가져와라
    print(keyword)
# reponse했을 시점의 데이터들임
```

* **naver검색어 환경상 1위가 두번나오는 오류 제거**

```python
real_rank_elements = dom.select("ol#realrank li.up") # li ~ 다음 li전까지임

for real_rank_element in real_rank_elements[:-1]: #네이버의 1위가 두번나오는 오류 제거해주기위해
    attrs = real_rank_element.select_one("a").attrs # a href~에 들어있는 모든 데이터 다가져옴
#    print(attrs)
    keyword = attrs.get("title")
    print(keyword)
```

* **title, link 뽑아오기**

```python
real_rank_elements = dom.select("ol#realrank li.up")

for real_rank_element in real_rank_elements[:-1]:
    attrs = real_rank_element.select_one("a").attrs 
    
    keyword = attrs.get("title")
    link = attrs.get('href')
    print(keyword, link)
    
```

## Homework2 : 정적인 사이트 하나 크롤링하기

### 예1) '신사동 오피스텔' 블로그 제목 크롤링하기
* 웹페이지 1 - 10페이지까지 크롤링

```python
import requests
from bs4 import BeautifulSoup

for page in range(1, 10):
    blog_post_id = 10 * (page - 1) + 1 
    print(blog_post_id)
    
    response = requests.get("https://search.naver.com/search.naver?where=post&sm=tab_pge&query=%EC%8B%A0%EC%82%AC%EB%8F%99+%EC%98%A4%ED%94%BC%EC%8A%A4%ED%85%94&start={page_num}".format(
            page_num=blog_post_id,
    ))
    
    dom = BeautifulSoup(response.content, "html.parser")
    post_elements = dom.select("ul.type01 li.sh_blog_top")
    for post_element in post_elements:
        title_element = post_element.select_one("a.sh_blog_title")
        title = title_element.attrs.get('title')
        print(title)
```
