### WHERE기능 모두 딕셔너리로 가능(편리함)

```python
MONGO_QUERY = {
    "item.deposit" : 1000,
    "item.rent" : 55,    
}

for item in collection.find(MONGO_QUERY):
    address = item.get('item').get('address1')
    deposit = item.get('item').get('deposit')
    rent = item.get('item').get('rent')
    print((deposit, rent, address))
```

```python
MONGO_QUERY = {
    "item.deposit" : {
        "$gte" : 1000,
        #g te => Greater Than Equal
        # gt =>
        # lte => Little Than Eual
        # lt
    },
    "item.rent" : 55,    
}

for item in collection.find(MONGO_QUERY):
    address = item.get('item').get('address1')
    deposit = item.get('item').get('deposit')
    rent = item.get('item').get('rent')
    print((deposit, rent, address))
```

```python
# 보증금 > 1000, 월세 > 50 이상 데이터
MONGO_QUERY = {
    "item.deposit" : {
        "$gte" : 1000,
        #g te => Greater Than Equal
        # gt =>
        # lte => Little Than Eual
        # lt
    },
    "item.rent" : {
        "$gte" : 50,
    },    
}

for item in collection.find(MONGO_QUERY):
    address = item.get('item').get('address1')
    deposit = item.get('item').get('deposit')
    rent = item.get('item').get('rent')
    print((deposit, rent, address))
```

```python
# 몽고디비는 RE 지원함 ($regex) => match(텍스트 찾거나 치환)
import pandas as pd

df = pd.DataFrame(columns=['address', 'deposit', 'rent'])

MONGO_QUERY = {
    "item.deposit" : {
        "$gte" : 1000,
        #g te => Greater Than Equal
        # gt =>
        # lte => Little Than Eual
        # lt
    },
    "item.rent" : {
        "$gte" : 50,
    },
    "item.address1": {
        "$regex" : "마포구",
    }
}

for item in collection.find(MONGO_QUERY):
    address = item.get('item').get('address1')
    deposit = item.get('item').get('deposit')
    rent = item.get('item').get('rent')
    print((deposit, rent, address))
    df.loc[len(df)] = [address, deposit, rent]
df
```
