# JOIN ON

```python
import pandas as pd
import MySQLdb

db = MySQLdb.connect(
    "db.fastcamp.us",
    "root",
    "dkstncks",
    "sakila",
    charset='utf8',
)

cursor = db.cursor()
```

```python
import time

start_time = time.clock()

SQL_QUERY = """
    SELECT *
    FROM customer
        JOIN address
        ON customer.address_id = address.address_id
    ;
"""

df = pd.read_sql(SQL_QUERY, db)

end_time = time.clock()

print("SELECT  JOIN: {time_spend}".format(time_spend = end_time - start_time))
```
