# WHERE LIKE

* Country 테이블에서 GovernmentForm에 "Republic"이라는 텍스트가 포함된 모든 row들을 뽑는다

** 1. SQL **

```python
import MySQLdb

db = MySQLdb.connect(
    "db.fastcamp.us",
    "root",
    "dkstncks",
    "world",
    charset='utf8',
)

cursor = db.cursor()

import pandas as pd
```

```python
SQL_QUERY = """
    SELECT Name, GovernmentForm
    FROM Country
    WHERE
        GovernmentForm LIKE "%Republic%"
    ;
"""

# % => 모든 텍스트 ( 없어도 되고 있어도 된다 )
# 앞에만 있으면 Republic으로 시작되는 텍스트를 뽑음
pd.read_sql(SQL_QUERY, db)
```

** 2. pandas **

```python
df = pd.read_sql("SELECT * FROM Country", db)
df
```

```python
df = df[ ["Name", "GovernmentForm"] ]
df
```

```python
df["GovernmentForm"].str.contains("Republic")
```

```python
df[df["GovernmentForm"].str.contains("Republic")]
```

# DISTINCT(unique한값들만뽑기)

** 1. SQL**

```python
# SQL 방법
SQL_QUERY = """
    SELECT DISTINCT(GovernmentForm)
    FROM Country;
"""

pd.read_sql(SQL_QUERY, db)
```
