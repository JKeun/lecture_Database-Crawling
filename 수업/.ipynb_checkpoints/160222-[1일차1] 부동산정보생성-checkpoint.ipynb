# 안수찬

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = ['address', 'phonenumber', 'deposit', 'rent',])
df
```

```python
# 밑에서 변수로 사용안할때 _로 함
import random
from faker import Faker
fake = Faker('ko_KR')

for _ in range(1000):
    deposit = random.choice(range(1, 20)) *500
    rent = random.choice(range(1, 20)) * 5
    phonenumber_list = [str(random.choice(range(0, 10))) for x in range(8)] # list comprehension
    phonenumber = '010' + "".join(phonenumber_list) #각각의list를 하나로 묶어주는작업
    phonenumber_processed = "-".join([
        # 1번 => "010"
            phonenumber[:3],
        # 2번 => "aaaa"
            phonenumber[3:-4],
        # 3번 => "bbbb")
            phonenumber[-4:]
        ]
     )
    address = fake.address().replace("\n", "")
    # print(deposit, rent, phonenumber_processed, address)
    
    df.loc[len(df)] = [address, phonenumber_processed, deposit, rent] # df.loc는 열하나씩 추가가능함 (ix와다른점)
```
