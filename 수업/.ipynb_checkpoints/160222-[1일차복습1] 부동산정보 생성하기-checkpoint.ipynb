```python
for _ in range(1000):
    deposit = random.choice(range(1, 20)) * 500
    rent = random.choice(range(1, 20)) * 5
    phonenumber_list = [str(random.choice(range(0, 10))) for x in range(8)] 
    # expected str 오류났으므로 str으로 타입지정
    # list comprehension
    phonenumber = '010' + "".join(phonenumber_list) # join : 각 리스트를 묶어줌
    print(phonenumber)
```

```python
for _ in range(1000):
    deposit = random.choice(range(1, 20)) * 500
    rent = random.choice(range(1, 20)) * 5
    phonenumber_list = [str(random.choice(range(0, 10))) for x in range(8)] 
    # expected str 오류났으므로 str으로 타입지정
    # list comprehension
    phonenumber = '010' + "".join(phonenumber_list) # join : 각 리스트를 묶어줌
    phonenumber_processed = "-".join([
            # 1번 => "010"
            phonenumber[:3],
            # 2번 => "aaaa"
            phonenumber[3:-4],
            # 3qjs => "bbbb"
            phonenumber[-4:]
        ])
    print(phonenumber_processed)
```

```python
for _ in range(1000):
    deposit = random.choice(range(1, 20)) * 500
    rent = random.choice(range(1, 20)) * 5
    phonenumber_list = [str(random.choice(range(0, 10))) for x in range(8)] 
    # expected str 오류났으므로 str으로 타입지정
    # list comprehension
    phonenumber = '010' + "".join(phonenumber_list) # join : 각 리스트를 묶어줌
    phonenumber_processed = "-".join([
            # 1번 => "010"
            phonenumber[:3],
            # 2번 => "aaaa"
            phonenumber[3:-4],
            # 3qjs => "bbbb"
            phonenumber[-4:]
        ])
    address = fake.address()
    print(address)
```

```python
for _ in range(1000):
    deposit = random.choice(range(1, 20)) * 500
    rent = random.choice(range(1, 20)) * 5
    phonenumber_list = [str(random.choice(range(0, 10))) for x in range(8)] 
    # expected str 오류났으므로 str으로 타입지정
    # list comprehension
    phonenumber = '010' + "".join(phonenumber_list) # join : 각 리스트를 묶어줌
    phonenumber_processed = "-".join([
            # 1번 => "010"
            phonenumber[:3],
            # 2번 => "aaaa"
            phonenumber[3:-4],
            # 3qjs => "bbbb"
            phonenumber[-4:]
        ])
    address = fake.address().replace("\n", "") # "\n" => ""으로 바꿔라
    print(address)
```

```python
for _ in range(1000):
    deposit = random.choice(range(1, 20)) * 500
    rent = random.choice(range(1, 20)) * 5
    phonenumber_list = [str(random.choice(range(0, 10))) for x in range(8)] 
    # expected str 오류났으므로 str으로 타입지정
    # list comprehension
    phonenumber = '010' + "".join(phonenumber_list) # join : 각 리스트를 묶어줌
    phonenumber_processed = "-".join([
            # 1번 => "010"
            phonenumber[:3],
            # 2번 => "aaaa"
            phonenumber[3:-4],
            # 3qjs => "bbbb"
            phonenumber[-4:]
        ])
    address = fake.address().replace("\n", "") # "\n" => ""으로 바꿔라
    # print(deposit, rent, phonenumber_processed, address)
    
    df.loc[len(df)] = [address, phonenumber_processed, deposit, rent]
    # df.loc는 열 한개 한개씩 추가 가능함
```

## index 제거하고 다시 넣기
