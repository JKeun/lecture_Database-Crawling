`$ pip install selenium `

```python
from selenium import webdriver
```

```python
driver = webdriver.Firefox() # local host
# driver = webdriver.PhantomJS() # server
```

```python
# visit following web page
driver.get("http://dabangapp.com") # path track ex) through Google Analytics
```

```python
from IPython.display import Image
```

```python
# screenshot of current web page
Image(driver.get_screenshot_as_base64(), width=100)
```

#### find "html element", take "action"(click, retrive...)
* send keys
* click

```python
# find_elements_by_css_selector: similar to "beautifulsoup dom"
input_element = driver.find_element_by_css_selector("div#home-component input.form-control") 
input_element
```

```python
input_element.send_keys(u"신사역")
```

```python
button_element = driver.find_element_by_css_selector("button.btn.btn-search")
button_element
```

```python
# click
button_element.click()
```

```python
# check by clikcing 'inspect' 
price_elements = driver.find_elements_by_css_selector("div.price.left")
len(price_elements)
```

```python
price_element = price_elements[0]
price_element.get_attribute("innerHTML") # innerhtml =  (similar to) .text
```

| Dom                      |Selenium                          |
|:------------------------:|:--------------------------------:|
|element.text              |element.get_attribute('innerHTML')|
|element.attrs.get('title')|element.get_attribute('title)     |

```python
for price_element in price_elements:
    print(price_element.get_attribute('innerHTML'))
```
