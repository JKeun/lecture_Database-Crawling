* 직방 api 웹페이지로 들어가 반복문 시행 (1000개 data 가져오기)

```python
BASE_URL = "https://api.zigbang.com/v1/items?detail=true"
```

```python
# API URL을 직접 생성해서
# response 가져오기
```

```python
zigbang_api_url = BASE_URL
```

```python
zigbang_api_url = zigbang_api_url + "&item_ids={item_id}".format(item_id=20000)
zigbang_api_url
```

```python
for i in range(200003, 200103):
    zigbang_api_url = zigbang_api_url + "&item_ids={item_id}".format(item_id=i)
```

```python
import requests
response = requests.get(zigbang_api_url)
```

```python
import json
json.loads(response.text)
```

```python
zigbang_dict = json.loads(requests.get(zigbang_api_url).text)
# 위의 코드와 같음
```

```python
zigbang_dict.get('count_agent')
```

```python
for item in zigbang_dict.get('items'):
    item_dict = item.get('item')
    #보증금, 월세, 휴대폰번호, 주소
    
print(item_dict.keys())
```

```python
item_dict.get('title')
```

```python
import pandas as pd
df = pd.DataFrame(columns = ['address', 'mobile', 'deposit', 'rent'])

for item in zigbang_dict.get('items'):
    item_dict = item.get('item')
    #보증금, 월세, 휴대폰번호, 주소
    deposit = item_dict.get('deposit')
    rent = item_dict.get('rent')
    mobile = item_dict.get('user_mobile')
    address = item_dict.get('agent_address1')
    #address1 = item_dict.get('address1')
    #address2 = item_dict.get('address2')
    #address = "{add1}, {add2}".format(add1 = address1 , add2 = address2,)
    # add1 => 동까지, add2 => 우편번호
    
    df.loc[len(df)]= [address, mobile, deposit, rent]
    # print(deposit, rent, mobile, address)
    
#print(item_dict.keys())
```

* csv파일로 저장하기
