## 0. 'world' DB에 접속

```python
import MySQLdb

db = MySQLdb.connect(
    "db.fastcamp.us",
    "root",
    "dkstncks",
    "world",
    charset='utf8',
)

cursor = db.cursor()
```

```python
import pandas as pd
```

## 1. 데이터 뽑기

** 1. SQL**

```python
pd.read_sql("""
SELECT *
FROM Country
WHERE Population > 100000000;
""", db)
```

** 2. pandas**

```python
df = pd.read_sql("SELECT * FROM Country;", db)
df
```

```python
df[df["Population"] > 100000000]
```

```python
df[df["Population"] > 100000000][["Name", "Population"]]
# df[['Name', 'Population]] 처럼 원래 df안에 리스트로 col_name을 집어넣는 것임
# 
```

## 2. 인구내림차순으로 뽑기

** 1. SQL**

```python
SQL_QUERY = """
    SELECT *
    FROM Country
    WHERE Population > 100000000
    ORDER BY Population DESC
"""
# DESC(ending) 내림차순
pd.read_sql(SQL_QUERY, db)
```

** 2. pandas**

```python
df = pd.read_sql("SELECT * FROM Country", db)
```

```python
new_df = df[df["Population"]> 100000000]
new_df
```

```python
new_df.sort_values('Population', ascending=False)
# 내림차순
```

## 3. 개수세기

**1. SQL**

```python
SQL_QUERY = """
    SELECT COUNT(*)
    FROM Country
    WHERE Population > 100000000
    ORDER BY Population DESC;
"""

pd.read_sql(SQL_QUERY, db)
```

** 2.pandas**

```python
len(new_df.sort_values('Population', ascending=False))
```
