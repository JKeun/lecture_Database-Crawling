# SQL&pandas Columns명 바꿔 뽑기 - world(db)
* SQL : SELECT Name (AS) 'Country Nmae'
* pandas : rename문

```python
import MySQLdb
import pandas as pd

db = MySQLdb.connect(
    'db.fastcamp.us',
    'root',
    'dkstncks',
    'world',
    charset='utf8',
)

cursor = db.cursor()
```

---
### SQL

```python
SQL_QUERY = """
   SELECT
       Name '국가명',
       Population '인구수'
   FROM
       Country
    ORDER BY 
        2 DESC        
    ;
"""

pd.read_sql(SQL_QUERY, db)
```

```python
SQL_QUERY = """
    SELECT 
        COUNT(*) '국가수'
    FROM 
        Country
    WHERE
        Population >= 40000000
    ;
"""

pd.read_sql(SQL_QUERY, db)
```

### pandas

```python
df = pd.read_sql("SELECT * FROM Country;", db)

new_df = df[['Name', 'Population']].rename(columns={
                                 "Name" : "국가명",
                                 "Population" : "인구수"
    })

new_df.sort_values(['인구수', '국가명'], ascending=[False, True])
```
