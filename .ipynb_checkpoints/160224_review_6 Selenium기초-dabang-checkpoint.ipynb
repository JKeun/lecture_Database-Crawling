```python
from selenium import webdriver
```

### dabang 홈피 크롤링하기

```python
# Localhost
driver = webdriver.Firefox()
# if use in server : drver = webdriver.PhantomJS()
```

```python
driver.get("http://fastcampus.co.kr")
```

```python
# Image는 아이파이썬 내장함수
# 현get_screenshot.. 현재 보고있는 페이지를 파일로 뽑는 드라이버 명령어임 
from IPython.display import Image
Image(driver.get_screenshot_as_base64())
```

```python
driver.get("http://dabangapp.com")
```

```python
# 관심지역 타이핑하기
# div id="home-component"
# class="form-control"
input_element = driver.find_element_by_css_selector(
    "div#home-component input.form-control")
```

```python
input_element
```

```python
input_element.send_keys(u"신사역")
```

```python
button_element = driver.find_element_by_css_selector(
"button.btn.btn-search")
```

```python
button_element.click()
```

```python
# 가격정보 element
price_elements = driver.find_elements_by_css_selector(
"div.price.left")
```

```python
len(price_elements)
```

```python
price_element = price_elements[0]
```

```python
price_element.get_attribute("innerHTML")
```

```python
# dom => element.txt
# selenium => element.get_attribute("innerHTML")

# dom = > element.attrs.get('title')
# selenium => element.get_attribute('title')
```

```python
for price_element in price_elements:
    print(price_element.get_attribute('innerHTML'))
```
