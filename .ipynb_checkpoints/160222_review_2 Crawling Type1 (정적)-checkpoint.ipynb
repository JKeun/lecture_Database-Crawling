```python
import pandas as pd

import requests
from bs4 import BeautifulSoup
```

## 1. 네이버 실검 크롤링

```python
response = requests.get("http://www.naver.com/")
dom = BeautifulSoup(response.content, "html.parser")

dom.select("ol#realrank li.up")
```

```python
df = pd.DataFrame(columns = ['keyword', 'link'])

response = requests.get("http://www.naver.com/")
#response.content
dom = BeautifulSoup(response.content, "html.parser")
#dom

real_rank_elements = dom.select("ol#realrank li.up")
#print(dom.select('ol#realrank li.up'))

for real_rank_element in real_rank_elements[:-1]:
    attrs = real_rank_element.select_one('a').attrs
    #print(real_rank_element.select_one('a'))
    
    keyword = attrs.get('title')
    #print(attrs.get('title'))
    link = attrs.get('href')
    
    df.loc[len(df)] = [keyword, link]

df
    
```

## 안보고 따라하기

```python
df = pd.DataFrame(columns = ['keyword', 'link',])

response = requests.get('http://naver.com/')
dom = BeautifulSoup(response.content, 'html.parser')

real_rank_elements = dom.select('li.up')

for real_rank_element in real_rank_elements[:-1]:
    attrs = real_rank_element.select_one('a').attrs
    
    title = attrs.get('title')
    link = attrs.get('href')
    
    df.loc[len(df)] = [title, link]
df
```

## 신사역 맛집 블로그 크롤링하기

```python
df = pd.DataFrame(columns = ['title', 'link'])

for page in range(5):
    page_num = 10*page +1
    print(page_num)
    
    response = requests.get("https://search.naver.com/search.naver?where=post&sm=tab_pge&query=%EC%8B%A0%EC%82%AC%EC%97%AD%20%EB%A7%9B%EC%A7%91&st=sim&date_option=0&date_from=&date_to=&dup_remove=1&post_blogurl=&post_blogurl_without=&srchby=all&nso=&ie=utf8&start={page}".format(page = page_num,
                                    ))
    
    dom = BeautifulSoup(response.content, 'html.parser')
    elements = dom.select("ul#elThumbnailResultArea li.sh_blog_top")
    
    for element in elements:
        blog_element = element.select_one('a.sh_blog_title').attrs
        title = blog_element.get('title')
        link = blog_element.get('href')
        
        df.loc[len(df)] = [title, link]
df
```
