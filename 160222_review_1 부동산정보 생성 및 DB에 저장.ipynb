# 부동산 매물 DataFrame 만들기
* feature : address, phonenumber, deposit, rent
* sample : 1000

```python
import pandas as pd

df = pd.DataFrame(columns = ['address', 'phonenumber', 'deposit', 'rent'])
df
```

* pip install fake-factory (주소랜덤생성 패키지)

```python
from faker import Faker
fake = Faker('ko_KR')
print(fake.name())
print(fake.address())
```

```python
import random
```

```python
df = pd.DataFrame(columns = ['address', 'phonenumber', 'deposit', 'rent'])

for _ in range(1000):
    address = fake.address().replace('\n', '')
    phonenumber_list = [str(random.choice(range(9))) for x in range(8)]
    deposit = random.choice(range(1, 20)) * 500
    rent = random.choice(range(1, 20)) * 5
    
    phonenumber = '010' + ''.join(phonenumber_list)
    phonenumber_processed = '-'.join([
            phonenumber[:3],
            phonenumber[3:-4],
            phonenumber[-4:]            
        ])
    
    df.loc[len(df)] = [address, phonenumber_processed, deposit, rent]
df    
    #print(phonenumber_processed)
```

# DB에 저장하기

```python
import MySQLdb

db = MySQLdb.connect(
    'db.fastcamp.us',
    'root',
    'dkstncks',
    'jkeun_sample',
    charset='utf8',
)

cursor = db.cursor()
```

```python
# data1000 테이블 생성하기

SQL_QUERY = """
    CREATE TABLE data1000 (
    address VARCHAR(255),
    phonenumber VARCHAR(255),
    deposit INT,
    rent INT
    )
;
"""
```

```python
for index, row in df.iterrows():
        
    print(row)
```

```python
for index, row in df.iterrows():
    address = row.get('address')
    phonenumber = row.get('phonenumber')
    deposit = row.get('deposit')
    rent = row.get('rent')
    
    SQL_QUERY = """
        INSERT INTO data1000 (address, phonenumber, deposit, rent)
        VALUES (
        '{address}',
        '{phonenumber}',
        '{deposit}',
        '{rent}'
        )
        ;    
    """.format(
        address = address,
        phonenumber = phonenumber,
        deposit = deposit,
        rent = rent,
    )
    
    cursor.execute(SQL_QUERY)
    db.commit()
```
