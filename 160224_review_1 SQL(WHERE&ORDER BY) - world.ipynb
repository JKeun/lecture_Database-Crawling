* 인구수 1억 이상인 국가를 뽑아라

```python
import MySQLdb
import pandas as pd

db = MySQLdb.connect(
    'db.fastcamp.us',
    'root',
    'dkstncks',
    'world',
    charset='utf8',
)

cursor = db.cursor()
```

---
### SQL(WHERE & ORDER BY DESC)

```python
SQL_QUERY = """
    SELECT *
    FROM Country
    WHERE Population > 100000000
    ORDER BY Population DESC
    ;
"""

pd.read_sql(SQL_QUERY, db)
```

### pandas

```python
df = pd.read_sql("SELECT * FROM Country;", db)

df[df['Population'] > 100000000].sort_values('Population', ascending=False)
```

---
### SQL(COUNT)

```python
SQL_QUERY = """
    SELECT COUNT(*)
    FROM Country
    WHERE Population > 100000000
    ORDER BY Population DESC
    ;
"""

pd.read_sql(SQL_QUERY, db)
```

### pandas

```python
df = pd.read_sql("SELECT * FROM Country;", db)

len(df[df['Population'] > 100000000])
```
