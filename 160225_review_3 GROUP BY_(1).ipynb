# GROUP BY _
* 월별 렌탈횟수(count) 높은 순으로뽑기 - rental table
* rental_date를 기준으로 뽑아야함

```python
import pandas as pd
import MySQLdb

db = MySQLdb.connect(
    "db.fastcamp.us",
    "root",
    "dkstncks",
    "sakila",
    charset='utf8',
)
```

### SQL

```python
SQL_QUERY = """
    SELECT
        LEFT(rental_date, 7) 'Monthly',
        COUNT(rental_id) 'Total Rental'
    FROM
        rental
    GROUP BY 1
    ORDER BY 2 DESC
    ;    
"""

pd.read_sql(SQL_QUERY, db)
```

### pandas
* pandas에서 미리 전처리를 해줘야함(groupby 기능이 없기때문에)

```python
rental_df = pd.read_sql("SELECT * FROM rental;", db)
rental_df
```

```python
# month라는 칼럼을 하나 추가한의미
rental_df['month']=rental_df['rental_date'].apply(
    lambda x: str(x)[:7])
rental_df
```

```python
rental_df.groupby('month').size()
# SQL에서의 COUNT는 pandas에서 size임
# row의 개수를 셈
```

```python
rental_df.groupby('month').count()
# count는 not null인 값들을 셈
```

### apply 사용법

```python
# 1번 방법 : lambda

# 2번 방법 : 함수를 만들어 씀

def get_address_id_falg(x):
    if x < 10:
        return 0
    else:
        return 1

rental_df['flat'] = rental_df['rental_id'].apply(get_address_id_falg)
```
