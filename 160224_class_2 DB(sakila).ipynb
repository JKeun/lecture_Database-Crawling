## 1. 데이터 뽑기
* rental_rate > 1, 내림차순
* title, rental_rate

** 1. SQL**

```python
SQL_QUERY = """
    SELECT title, rental_rate
    FROM film
    WHERE rental_rate >= 1
    ORDER BY rental_rate DESC;  # rental_rate 기준
"""

pd.read_sql(SQL_QUERY, db)
```

```python
SQL_QUERY = """
    SELECT title, rental_rate
    FROM film
    WHERE rental_rate >= 1
    ORDER BY
        rental_rate DESC, # 제1기준
        title ASC;        # 제2기준
"""

pd.read_sql(SQL_QUERY, db)
```

** 2. pandas**

```python
df = pd.read_sql("SELECT * FROM film;", db)
df
```

```python
df[df["rental_rate"]> 1][["title", "rental_rate"]].sort_values('rental_rate',
                                                                ascending=False)
```

```python
# 위와 같음
df[    ['title', 'rental_rate']    ][  df['rental_rate'] > 1.00    ].sort_values(
    "rental_rate", ascending = False)

# df[['title', 'rental_rate']]는 데이터를 받아오는 명령이기때문에 리스트로 묶고
# df['rental_rate']는 boolean Series를 받기위한것이기 때문에 리스트로 묶지않는다
```

```python
df[    ['title', 'rental_rate']    ][  df['rental_rate'] > 1.00    ].sort_values(
    ["rental_rate", "title"], ascending = [False, True])
```
