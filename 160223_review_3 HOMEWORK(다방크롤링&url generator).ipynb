### URL Generator사용하여 1000개 데이터 크롤링하기

```python
df = pd.DataFrame(columns=['name', 'deposit', 'rent', 'address', 'floor' ])

for page in range(1, 11):
    response = requests.get("http://www.dabangapp.com/api/2/room/list/search?page={i}&filters=%7B%22deposit-range%22%3A%5B0%2C999999%5D%2C%22price-range%22%3A%5B0%2C999999%5D%2C%22room-type%22%3A%5B0%2C1%2C2%2C3%2C4%2C5%5D%2C%22deal-type%22%3A%5B0%2C1%5D%2C%22location%22%3A%5B%5B126.65657267485665%2C37.505146062104686%5D%2C%5B126.90719828520821%2C37.560414154305995%5D%5D%7D&_=1456933964300".format(i=page))
    dabang_dict = json.loads(response.text)
    rooms = dabang_dict.get('rooms')
    
    for room in rooms:
        address = room.get('address')
        deposit = room.get('deposit')
        rent = room.get('price')
        name = room.get('name')
        floor = room.get('floor')
    
        df.loc[len(df)] = [name, deposit, rent, address, floor]        
```
