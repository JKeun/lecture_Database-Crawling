```python
# SELECT FROM WHERE 를 이용한 무식한 방법

start_timen = time.clock()

SQL_QUERY = """
    SELECT *
    FROM
        customer, address
    WHERE
        customer.address_id = address.address_id
    ;
"""

df = pd.read_sql(SQL_QUERY, db)

end_time = time.clock()

print("SELECT WHERE: {time_spend}".format(time_spend = end_time - start_time))
```

```python
len(df)
```

### JOIN으로 뽑는 경우

```python
# SELECT JOIN 을 이용한 유식한 방법

start_time = time.clock()

SQL_QUERY = """
        JOIN address
        ON customer.address_id = address.address_id
    ;
"""

df = pd.read_sql(SQL_QUERY, db)

end_time = time.clock()

print("SELECT WHERE: {time_spend}".format(time_spend = end_time - start_time))
```

```python
len(df)
```
